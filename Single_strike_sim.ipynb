```python
from force_model_functions import *
```

### Environment parameters:

```python
#Time-step
del_t = 0.01

#Friction coefficient between slider and surface:
mu1 = 0.03

#Friction coefficient between gripper and slider:
mu2 = 0.1

#Block dimensions: (Shape dependent)
length = 0.3
breadth = 0.1
height = 0.1

#Mass of block in kg:
mass = 0.2
```

```python
params = {}

params['Fr'] = mu1 * mass * 9.81
params['mu1'] = mu1
params['mu2'] = mu2
params['length'] = length
params['breadth'] = breadth
params['mass'] = mass

#Shape dependent:
params['Iz'] = (mass * (length**2 + breadth**2)) / 12
```

### Generate a push:

```python
initial_pos = np.array([0.5, 0.5, 0.0])
initial_vel = np.array([0., 0., 0.])

force = np.array([0., 20.])
phi = -30 * np.pi/180
```

```python
acc, vel, pos = simulate(force, phi, initial_pos, initial_vel, del_t, params)
```

```python
np.save("pos.npy", pos)
```
