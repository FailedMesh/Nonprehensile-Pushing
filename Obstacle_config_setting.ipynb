### Create scene:

```python
obstacle_locations = np.zeros((6, 4))  #Scenes x num obstacles x pose

bounding_threshold = 0.2
shapes = ['cube', 'cuboid', 'cylinder']

used_locs = np.zeros((0, 2))

for i in range(obstacle_locations.shape[0]): 
    valid_pos = False
    while not valid_pos:
        location = np.random.uniform(low = [0.15, -0.65],
                                    high = [0.65, -0.15])
        valid_pos = True
        for j in range(used_locs.shape[0]):
            if np.linalg.norm(location - used_locs[j]) <= bounding_threshold:
                valid_pos = False
        if valid_pos:
            used_locs = np.concatenate([used_locs, np.array([location])], axis = 0)
    obstacle_locations[i, :2] = location.copy()
    
    obstacle_locations[i, 3] = np.random.randint(3)
    
    obstacle_locations[i, 2] = np.random.uniform(0, 2*np.pi)
    
print(used_locs)
print(obstacle_locations)

#plt.scatter(obstacle_locations[0, :, 0], obstacle_locations[0, :, 1], color = 'green', marker = 'o', label = 'Obstacles')
plt.scatter(0.0, -0.8, color = 'black', label = 'Start')
plt.scatter(0.8, 0.0, color = 'red', label = 'Goal')


for i in range(obstacle_locations.shape[0]):
    
    obs_type = shapes[int(obstacle_locations[i, 3])]
    
    if obs_type == 'cylinder':
        obstacle = plt.Circle((obstacle_locations[i, 0], obstacle_locations[i, 1]), color = 'green',
                              radius = 0.025)
        
    elif obs_type == 'cube':
        obstacle = plt.Rectangle((obstacle_locations[i, 0] - 0.05/2, obstacle_locations[i, 1] - 0.05/2), 
                                0.05, 
                                0.05, 
                                color = 'green', 
                                angle = obstacle_locations[i, 2] * 180/np.pi, 
                                rotation_point = 'center')
        
    elif obs_type == 'cuboid':
        obstacle = plt.Rectangle((obstacle_locations[i, 0] - 0.25/2, obstacle_locations[i, 1] - 0.05/2), 
                                0.25, 
                                0.05, 
                                color = 'green', 
                                angle = obstacle_locations[i, 2] * 180/np.pi, 
                                rotation_point = 'center')
    
    plt.gca().add_patch(obstacle)
    
plt.xlim([-1, 1])
plt.ylim([-1, 1])

plt.show()

_ = input("Do you want to save?")

np.save('obstacle_locations/complex_scene1.npy', obstacle_locations)
    

# for i in range(obs.shape[0]):
    
    
#     obstacle = plt.Rectangle((obs[i, 0] - obs_l/2, obs[i, 1] - obs_b/2), 
#                                 obs_l, 
#                                 obs_b, 
#                                 color = 'green', 
#                                 angle = 90, #obs[i, 2] * 180/np.pi, 
#                                 rotation_point = 'center')

#     plt.gca().add_patch(obstacle)
```
